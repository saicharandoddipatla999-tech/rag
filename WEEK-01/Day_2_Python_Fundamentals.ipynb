**3. Even or Odd**

**Problem Statement:**

*Take a number as input and check whether it is Even or Odd.*

```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

**4. Positive, Negative, or Zero**

**Problem Statement:**

*Take a number as input and determine whether it is:*

```python
num = int(input("Enter a number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

**8. Multiplication Table**

**Problem Statement:**

*Take a number as input and generate its multiplication table.*

```python
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

**9. Sum of Numbers from 1 to N**

**Problem Statement:**

*Take a number N and find the sum from 1 to N.*

```python
n = int(input("Enter N: "))

total = 0

for i in range(1, n + 1):
    total += i

print("Sum =", total)
```

   **10. Count Even and Odd Numbers**

**Problem Statement:**

**Take a number N and count:**

*● Total Even Numbers*

● Total Odd Numbers
*between 1 and N.*

```python
n = int(input("Enter N: "))

even_count = 0
odd_count = 0

for i in range(1, n + 1):
    if i % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

print("Even Count =", even_count)
print("Odd Count =", odd_count)
```
